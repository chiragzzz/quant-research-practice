```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

 Download NIFTY & BANKNIFTY Data

```python
nifty = yf.download("^NSEI",start = "2023-01-01", end = "2025-06-21", auto_adjust = False)
banknifty = yf.download("^NSEBANK",start = "2023-01-01", end = "2025-06-21", auto_adjust = False)

```

```python
nifty.head()
```

```python
banknifty.head()
```

```python
if isinstance(nifty.columns, pd.MultiIndex):
    nifty.columns = [f"{col[0]}_{col[1]}" for col in nifty.columns]
    close_colnifty = [col for col in nifty.columns if col.startswith("Close")][0]
       
else:
    close_colnifty = "Close"


##############################

if isinstance(banknifty.columns, pd.MultiIndex):
    banknifty.columns = [f"{col[0]}_{col[1]}" for col in banknifty.columns]
    close_colbank = [col for col in banknifty.columns if col.startswith("Close")][0]
else:
    close_colbank = "Close"
        
```

```python
nifty.head()
```

```python
banknifty.head()
```

```python
nifty = nifty[[close_colnifty]].rename(columns={close_colnifty: "NIFTY"})
banknifty = banknifty[[close_colbank]].rename(columns = {close_colbank : "BANKNIFTY"})
```

```python
nifty.head()
```

```python
banknifty.head()
```

```python
nifty = nifty.reset_index()
banknifty = banknifty.reset_index()
```

```python
nifty.head()
```

```python
banknifty.head()
```

 Merge both by Date

```python
df = pd.merge(nifty, banknifty, on = "Date")
```

```python
df.head()
```

Normalize Prices

```python
df["NIFTY_norm"] = df["NIFTY"]/df["NIFTY"].iloc[0]*100
df["BANKNIFTY_norm"] = df["BANKNIFTY"] / df["BANKNIFTY"].iloc[0]*100
```

```python
df.head()
```

Plot Comparative Performance

```python
plt.figure(figsize= (12, 6))
plt.plot(df["Date"], df["NIFTY_norm"], label = "Nifty", linewidth = 2, color = "green", alpha = 0.8)
plt.plot(df["Date"], df["BANKNIFTY_norm"], label = "BankNifty", linewidth = 2, color = "blue", alpha = 0.8 )

plt.title("NIFTY v/s BANKNIFTY - Normalized Performance (2023-2025)")

plt.xlabel("Date")
plt.ylabel("Indexed Price (Base = 100)")
plt.grid(True)

plt.legend()
plt.tight_layout()

plt.savefig("nifty_banknifty_comparision.png")
plt.show()
```

```python
df.to_csv("nifty_banknifty_normalized.csv", index = False)
```
